### Exploratory Data Analysis

```python
crashes_by_year = df.groupby('YEAR').size()
plt.plot(crashes_by_year.index, crashes_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of crashes')
plt.title('Number of crashes in Maryland (2016-2023)')
plt.show()
```

```python
# setting index of dataframe to accident date
df.set_index('datetime', inplace=True)
```

```python
# Monthly crashes report
df.resample('M').size().plot(legend=False)
plt.title('Number of Monthly Crashes')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
df_ts_resampled = df.resample('D').size() # Resample the data to daily frequency and count the number of crashes per day
if df_ts_resampled.isna().sum() > 0:
    df_ts_resampled = df_ts_resampled.fillna(0) # Replace missing values with 0
df_ts_rolling = df_ts_resampled.rolling(window=30).mean() # Calculate the 30-day rolling mean
fig, ax = plt.subplots(figsize=(12, 5))
plt.plot(df_ts_rolling)
plt.title('30-Day Rolling Mean of Daily Vehicle Crashes in Maryland')
plt.xlabel('Date')
plt.ylabel('Number of Crashes')
plt.show()
```

```python
# Resample the data by month and count the number of crashes per month
crashes_by_month = df.resample('M').size()

# Decompose the time series to analyze trends and seasonal patterns
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(crashes_by_month, model='additive')

# Plot the decomposed time series
fig, axes = plt.subplots(ncols=1, nrows=4, sharex=True, figsize=(10,8))
result.observed.plot(ax=axes[0], legend=False)
axes[0].set_ylabel('Observed')
result.trend.plot(ax=axes[1], legend=False)
axes[1].set_ylabel('Trend')
result.seasonal.plot(ax=axes[2], legend=False)
axes[2].set_ylabel('Seasonal')
result.resid.plot(ax=axes[3], legend=False)
axes[3].set_ylabel('Residual')
plt.show()
```

```python
# value counts of harm events description
# df.HARM_EVENT_DESC1.value_counts()
import seaborn as sns
counts = df['HARM_EVENT_DESC1'].value_counts().sort_values(ascending=False)
sns.countplot(x='HARM_EVENT_DESC1',palette = 'deep', data=df, order=counts.index)
plt.gcf().set_size_inches(15,8)
plt.xticks(rotation='vertical')
plt.title('Distribution of the HARM_EVENT_DESC')
plt.xlabel('Harm Event Description')
plt.ylabel('Count')
plt.show()
```
